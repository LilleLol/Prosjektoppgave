Siden $v_{max}$, $u_{max}$, $p$ og $u \geq 0$ endres ikke fortegnet, og gir at $\dfrac{dv}{du} \leq 0$

**Krav 4: v(0) = v_max**

$$v(0) = v_{max}\left(1 - \left(\frac{0}{u_{max}}\right)^p\right) = v_{max} \cdot (1 - 0) = v_{max}$$

**Krav 5: v(u_max) = 0**

$$v(u_{max}) = v_{max}\left(1 - \frac{u_{max}}{u_{max}}\right)^p = v_{max} \cdot (1 - 1) = 0$$

### c) De Deriverte

Derivert | Utregning | Enhet |

**$\dfrac{dv}{du}$** beskriver hvor sensitiv hastigheten er for endringer i tetthet. Dette kan brukes til å forstå når fri flyt bryter sammen til kø. Siden $\dfrac{dv}{du} \leq 0$ alltid, vil økt tetthet aldri gi økt hastighet.

**$\dfrac{\partial u}{\partial t}$** beskriver om trafikken bygger seg opp eller løser seg opp i et fast punkt over tid. Positiv verdi betyr at køen vokser, negativ verdi betyr at køen løser seg opp. Altså trafikkflyten i et fast punkt.

**$\dfrac{\partial u}{\partial x}$** beskriver hvordan tettheten varierer langs veien. En negativ verdi er et tegn på at det kan være kø foran. En positiv verdi betyr at veien åpner seg fremover.

**$\dfrac{\partial v}{\partial t}$** beskriver hvordan hastighetsnivået i et fast punkt på veien endrer seg over tid.

**$\dfrac{\partial v}{\partial x}$** beskriver om bilene foran kjører fortere eller saktere enn den bestemte bilen. En negativ verdi betyr at bilene foran bremser ned. Enheten kan tolkes som en "trafikk-frekvens" som angir hvor raskt hastighetsforholdene varierer langs veien.

### d) Utregning av fluksen

Fluks er hvor mange biler som passerer en strekning på en hvis tid, oppgitt i $\text[{biler/tid}]$

Begynner med å finne mengde biler fra tettheten for å definere biler og tiden

$u(x,t)$ = $[\text{biler}/\text{lengde}]$, setter inn en vilkårlig lengde $\Delta x$, hvor

$\;u(x,t)\cdot\Delta x = [\text{biler}]$ 

Definerer tiden fra farten gitt $v(u(x,t))$:
$$v(u(x,t)) = \frac{\Delta x}{\Delta t} \implies \Delta t = \dfrac{\Delta x}{v(u(x,t))}$$

Setter dette sammen
$$J = \frac{\text{Biler}}{\text{Tid}} \implies \frac{u(x,t) \cdot \Delta x}{\Delta t} \implies \frac{u(x,t) \cdot \Delta x}{\dfrac{\Delta x}{v(u(x,t))}} \implies u(x,t) \cdot v(u(x,t))$$

Enhetssjekk:

$$u(x,t) \cdot v(u(x,t)) \implies \frac{\text{biler}}{\text{lengde}} \cdot \frac{\text{lengde}}{\text{tid}} = \frac{\text{biler}}{\text{tid}} \checkmark$$

### e) Maksimal fluks 


i) 

For å finne maksimal fluks må vi finne hvilken $u$ som gir $J(u) = J_{\max}$.
Vi vet at $J(u) \geq 0$ for alle $u$, og at $J(u)$ har nullpunkter i $J(0) = 0$ og 
$J(u_{\max}) = 0$ siden $v(u_{\max}) = 0$. Siden $v$ ikke kan stige når $u$ øker, 
vil $J(u)$ kun ha ett ekstremalpunkt, og dette er et toppunkt. Derfor må $u$-verdien vi får fra $\dfrac{dJ}{du}=0$ være lik $J(u)=J_{\max}$.

Vi finner $\dfrac{dJ}{du}$:

$$J(u) = u \cdot v_{\max}\left(1 - \frac{u}{u_{\max}}\right)$$

$$\frac{dJ}{du} = \frac{d}{du}\left(u \cdot v_{\max}\left(1 - \frac{u}{u_{\max}}\right)\right)$$

Vi bruker produktregelen og får:

$$\frac{dJ}{du} = v_{\max}\left(1 - \frac{2u}{u_{\max}}\right)$$

For at $\dfrac{dJ}{du} = 0$ må minst ett av leddene være 0, og vi vet fra før at $v_{\max} \neq 0$:

$$1 - \frac{2u}{u_{\max}} = 0 \implies \frac{2u}{u_{\max}} = 1 \implies u^* = \frac{u_{\max}}{2}$$

Hastigheten ved maksimal fluks da:

$$v\!\left(\frac{u_{\max}}{2}\right) = v_{\max}\!\left(1 - \frac{\dfrac{u_{\max}}{2}}{u_{\max}}\right) = \frac{1}{2}v_{\max}$$

Den maksimale fluksen blir:

$$J_{\max} = J\!\left(\frac{u_{\max}}{2}\right) = \frac{u_{\max}}{2} \cdot \frac{v_{\max}}{2} = \frac{u_{\max} \cdot v_{\max}}{4}$$

ii) 

Vi setter $v(u) = v_{\max}\!\left(1 - \left(\dfrac{u}{u_{\max}}\right)^{\!P}\right)$ og antar $P \in \mathbb{N}$.
Vi må fortsatt finne $\dfrac{dJ}{du} = 0$:

$$\frac{dJ}{du} = \frac{d}{du}\left(u \cdot v_{\max}\left(1 - \left(\frac{u}{u_{\max}}\right)^{\!P}\right)\right) = \frac{d}{du}\left(v_{\max}\left(u - \frac{u^{P+1}}{u_{\max}^P}\right)\right)$$

$$\frac{dJ}{du} = v_{\max}\left(1 - \frac{(P+1)\,u^P}{u_{\max}^P}\right)$$

Vi løser $\dfrac{dJ}{du} = 0$, og siden $v_{\max} \neq 0$:

$$1 - \frac{(P+1)\,u^P}{u_{\max}^P} = 0 \implies u^* = u_{\max}\!\left(\frac{1}{P+1}\right)^{\!1/P}$$

Setter inn i $v(u)$:

$$v(u^*) = v_{\max}\!\left(1 - \left(\frac{u_{\max}\left(\frac{1}{P+1}\right)^{1/P}}{u_{\max}}\right)^{\!P}\right) = v_{\max}\!\left(1 - \frac{1}{P+1}\right) = v_{\max}\cdot\frac{P}{P+1}$$

Den maksimale fluksen blir:

$$J_{\max} = u^* \cdot v(u^*) = u_{\max}\!\left(\frac{1}{P+1}\right)^{\!1/P} \cdot v_{\max}\cdot\frac{P}{P+1}$$

Vi tester med $P = 1$ ettersom dette blir likt som oppgave i):

$$v(u^*) = v_{\max}\cdot\frac{1}{2} = \frac{v_{\max}}{2} $$

Dette stemmer med resultatet vi fikk i fra oppgave i).

### f) Differensiallikningen

Vi sjekker at begge sider av likninen har samme måleenhet.

Venstre side $\frac{\partial u(x,t)}{\partial t}$, der $u$ har enhet $[\text{biler}/m]$ og $t$ har enhet $[s]$:

$$\frac{\partial u}{\partial t} \sim \frac{\text{biler}/m}{s} = \frac{\text{biler}}{m \cdot s}$$

Høyre side $\frac{\partial J(x,t)}{\partial x}$, der $J = u \cdot v$ og $x$ har enhet $[m]$.

Først finner vi enheten til $J$, der $v$ har enhet $[m/s]$:

$$J \sim \frac{\text{biler}}{m} \cdot \frac{m}{s} = \frac{\text{biler}}{s}$$

Deretter deriverer vi med hensyn på $x \ [m]$:

$$\frac{\partial J}{\partial x} \sim \frac{\text{biler}/s}{m} = \frac{\text{biler}}{m \cdot s}$$

### d) Bevegelsen til en enkelt bil

Vi starter med å finne tettheten $u(x, t)$ i det området bilen befinner seg. Når lyset blir grønt dannes en "bølge" av akselerasjon som sprer seg bakover i køen. 

Før bølgen intreffer har vi en stillestående kø:

$u = u_{\max}$ for $x < -v_{\max}t$

I akselerasjonssonen får vi en overgang

$u=\dfrac{u_{\max}}{2}\!\left(1 - \dfrac{x}{v_{\max}\,t}\right)$ for $-v_{\max}\,t \leq x \leq v_{\max}\,t \quad$


Etter akselerasjonssonen vil bilene bevege seg med fri flyt

$u=0$ for $x > v_{\max}t$


Setter inn i formel for å finne hastigheten i akselerasjonssonen

$$v(x,t) = v_{\max}\!\left(1 - \tfrac{1}{2}\!\left(1 - \frac{x}{v_{\max}\,t}\right)\right) = \frac{1}{2}\!\left(v_{\max} + \frac{x}{t}\right)$$


Bilen står stille ved $P(0) = -1000$ helt til bølgen fra trafikklyset når den. Med verdien $v_{max}$ = $50 km/t$ = $13,9 m/s$ vil akselerasjonsbølgen nå bilen etter:

$$t_{\text{start}} = \frac{P(0)}{-v_{\max}} = \frac{-1000}{-13{,}9} \approx 71{,}9\,\text{s}$$

for $0 < t < t_{\text{start}}$: $V(t) = 0$ og $P(t) = -1000$. Når $t > t_{\text{start}}$ befinner 
bilen seg i akselerasjonssonen. Setter inn bilens posisjon $P(t)$ for $x$:

$$\frac{dP}{dt} = \frac{1}{2}\left(v_{\max} + \frac{x}{t}\right) \implies \frac{dP}{dt} = \frac{1}{2}v_{\max} + \frac{P(t)}{2t}$$

Dette er en lineær førsteordens differensiallikning på formen:

$$\frac{dP}{dt} - \frac{1}{2t}P = \frac{1}{2}v_{\max}$$

Løser med integrerende faktor $p(t)$:

$$p(t) = e^{\int -\frac{1}{2t}\,dt} = e^{-\frac{1}{2}\ln(t)} = t^{-\frac{1}{2}} = \frac{1}{\sqrt{t}}$$

Multipliserer hele likningen med $\dfrac{1}{\sqrt{t}}$:

$$\frac{dP}{dt}\cdot\frac{1}{\sqrt{t}} - \frac{1}{2t\sqrt{t}}P = \frac{1}{2\sqrt{t}}v_{\max} \implies \frac{d}{dt}\!\left(\frac{P}{\sqrt{t}}\right) = \frac{v_{\max}}{2\sqrt{t}}$$

Integrerer begge sider med hensyn på $t$:

$$\frac{P}{\sqrt{t}} = \int \frac{v_{\max}}{2\sqrt{t}}\,dt = v_{\max}\cdot\sqrt{t} + C$$

Ganger med $\sqrt{t}$ og får den generelle løsningen:

$$P(t) = v_{\max}\,t + C\sqrt{t}$$

Bestemmer integrasjonskonstanten ved å sette inn $P(t_{\text{start}}) = P_0$, med verdiene 
$P_0 = -1000$, $v_{\max} = 13{,}9$ og $t_{\text{start}} = 71{,}9$:

$$P_0 = v_{\max}\cdot t_{\text{start}} + C\sqrt{t_{\text{start}}}$$

Siden $v_{\max}\cdot t_{\text{start}} = |P_0|$ (distansen bølgen har reist), får vi:

$$-1000 = 1000 + C\sqrt{71{,}9} \implies C = -2\sqrt{v_{\max}|P_0|} = \frac{-2000}{\sqrt{71{,}9}} \approx -235{,}9$$

Setter inn $C = -2\sqrt{v_{\max}|P_0|}$ og får posisjonen:

$$P(t) = v_{\max}\,t - 2\sqrt{v_{\max}\cdot|P(0)|\cdot t}$$

Hastighet ($V = P'$):

$$V(t) = v_{\max} - \sqrt{\frac{v_{\max}\cdot|P_0|}{t}}$$



### e) Hastigheten til en kø

Når trafikklyset skifter til grønt ved $x=0$, begynner den fremste bilen å bevege seg. Siden informasjonen om det grønne lyset beveger seg med en konstant hastighet bakover gjennom køen, blir farten $-v_{max}$. Dette ser vi også på fluksen. 

Vi vet at formelen for fluksen er 

$$J=u∙v(u)= v_{max} (u-\frac{u^2}{u_{max}} )$$

noe som gir oss $$\frac{dJ}{du}=v_{max} (1-\frac{2u}{u_{max}} )$$

Dette betyr at når $u=u_{max}$ vil endringen av fluksen være lik 

$$J' (u_{max} )=v_{max} (1-2)=(-v)_{max}$$

Startpunktet er $x=0$ ved $t=0$, posisjonen $K(t)$ blir da: 
$K(t)=startposisjon+(fart∙tid)   →  K(t)=0+(-13,9∙t)$
Alle bilene er i bevegelse når bilen i posisjonen $K(t)=-1000$ starter å akselerere:

$$-13,9∙t=-1000$$

$$t=\frac{-1000}{-13,9}≈71,9$$

Når vi setter $P=2$ får vi formelen for endring av fluks lik 
$$\frac{dJ}{du}=v_{max} (1-\frac{(2+1) u^2}{u_{max}^2 })=v_{max} (1-\frac{3u^2}{u_{max}^2 })$$

Dette gir oss $$J' (u_{max} )=v_{max} (1-3)=13,9m/s∙(-2)=-27,8m/s$$

Altså vil siste bil starte å akselerere etter $t=\frac{-1000}{-27,8}≈36$ sekunder. 
Når vi setter $P=5$ ender vi opp med en ganske lik formel. Vi får
$$J' (u_{max} )=v_{max} (1-\frac{(5+1) u^5}{u_{max}^5 })=v_{max} (-5)=-69,5m/s$$

Siste bil vil altså starte å akselerere etter $t=\frac{-1000}{-69,5}≈14$ sekunder. 



### f) Valg av parameter $p$

Parameteren $p$ i hastighetsfunksjonen beskriver hvordan sjåfører reagerer på endring i tetthet.

For $p = 1$ får vi en lineær modell hvor sjåførene reduserer farten jevnt og proposjonalt med tettheten.

Ved $p > 1$ endres kurven slik at hastigheten holder seg høyere lenger enn i den lineære modellen, selv når tettheten øker. Jo høyere $p$, desto mer optimistisk kjøreatferd. Farten holdes nær $v_{\max}$ lenger, men fallet mot $v=0$ blir tilsvarende brattere.

Dette kan vi se dersom vi ser på formelen for $v(u)$. 

$$v(u)=v_{max}(1-( \frac{u}{u_{max}} )^P )$$

Her ser vi at når P går mot uendelig vil $v(u)$ gå mot 0 uansett $u$, unntatt $u=u_{max}$

### a)

Randverdiproblemet:

$$u_{xx}(x) = \cos(\pi x), \quad -1 \leq x \leq 1, \quad u(-1)=0,\; u(1)=2$$

$$u_x(x) = \frac{1}{\pi}\sin(\pi x) + C_1$$

Randsbetingelse $x = -1$:

$$u(-1) = -\frac{1}{\pi^2}\cos(-\pi) - C_1 + C_2 = \frac{1}{\pi^2} - C_1 + C_2 = 0$$

Randsbetingelse $x = 1$:

$$u(1) = -\frac{1}{\pi^2}\cos(\pi) + C_1 + C_2 = \frac{1}{\pi^2} + C_1 + C_2 = 2$$

Løser systemet: Fra likning 1 og 2:

$$\frac{2}{\pi^2} + 2C_2 = 2 \implies C_2 = 1 - \frac{1}{\pi^2}$$

$$C_1 = 1$$

Analytisk løsning:

$${u(x) = -\frac{1}{\pi^2}\cos(\pi x) + x + 1 - \frac{1}{\pi^2}}$$

### b)

Vi diskretiserer likningen og erstatter den andrederiverte $u_{xx}$ med approksimasjonen senterdifferanse:

$$u_{xx}(x) = \cos(\pi x), \quad -1 \leq x \leq 1$$

$$u_{xx}(x_i) \approx \frac{u_{i+1} - 2u_i + u_{i-1}}{h^2}$$

hvor $h$ er steglengden og $u_i$ er den numeriske verdien i punkt $x_i$. Likningen blir da:

$$\frac{u_{i+1} - 2u_i + u_{i-1}}{h^2} = \cos(\pi x_i)$$

Vi deler intervallet $[-1, 1]$ i $N$ delintervaller:

$$h = \frac{1-(-1)}{N} = \frac{2}{N}$$

$$u_0 = 0, \qquad u_N = 2$$

Dette gir et lineært likningssystem på formen $Au = b$.



Den numeriske løsningen ser ut til å konvergere mot den analytiske når $h \to 0$.



## Oppgave 4: Varmeligning, 1D (5 poeng)

Når $u_t = 0$ slutter systemet å endre seg:

$$u_{xx} = \cos(\pi x) \implies u_{xx} - \cos(\pi x) = 0$$

$$u_t = u_{xx} - \cos(\pi x), \quad x \in [-1, 1],\ t \geq 0$$

Den stasjonære løsningen fra oppgave 3 er likevektstilstanden for den tidsavhengige varmelikningen. Hvis $u_{xx} > \cos(\pi x)$ øker temperaturen, og ved $u_{xx} < \cos(\pi x)$ synker den. Systemet justerer seg helt til $u_{xx} = \cos(\pi x)$:

$$u_t = \cos(\pi x) - \cos(\pi x) = 0$$

Diskretisering av området:

$$x_i = -1 + i\Delta x, \quad i \in [0, N]$$

Romderiverte med senterdifferanse:

$$u_{xx}(x_i, t^n) \approx \frac{u_{i+1}^n - 2u_i^n + u_{i-1}^n}{(\Delta x)^2}$$

Tidsderiverte med forover differanse:

$$u_t(x_i, t^n) \approx \frac{u_i^{n+1} - u_i^n}{\Delta t}$$

Setter inn i likningen:

$$\frac{u_i^{n+1} - u_i^n}{\Delta t} = \frac{u_{i+1}^n - 2u_i^n + u_{i-1}^n}{(\Delta x)^2} - \cos(\pi x_i)$$

Oppdateringsformel:

$$u_i^{n+1} = u_i^n + \frac{\Delta t}{(\Delta x)^2}\left(u_{i+1}^n - 2u_i^n + u_{i-1}^n\right) - \Delta t\cos(\pi x_i)$$

Randsbetingelser for alle $n$ (oppdateres ikke, brukes direkte):

$$u_0^n = 0, \qquad u_N^n = 2$$

Initialbetingelse for alle $i$:

$$u_i^0 = 1 + x_i + 5\sin(\pi x_i)$$

Stabilitetskrav:

$$\frac{\Delta t}{(\Delta x)^2} \leq \frac{1}{2} \implies \Delta t \leq \frac{(\Delta x)^2}{2}$$

## Oppgave 5: Poissonligning, 2D (5 poeng)

Siden $f = 0$ løser vi egentlig Laplace-likningen for de gitte betingelsene. Bruker differansemetoden for å løse numerisk. Starter med diskretisering i området:

$$x \in [-5, 5], \quad y \in [0, 2], \quad u_{i,j} \approx u(x_i, y_j)$$

Vi deler inn i $N$ intervaller med steglengde $\Delta x = \Delta y = h$, slik at:

$$x_i = -5 + ih, \quad y_j = jh$$

Approksimasjon med senterdifferanse av de deriverte:

$$u_{xx} \approx \frac{u_{i+1,j} - 2u_{i,j} + u_{i-1,j}}{h^2}, \qquad u_{yy} \approx \frac{u_{i,j+1} - 2u_{i,j} + u_{i,j-1}}{h^2}$$

Setter dette inn i $u_{xx} + u_{yy} = 0$, og velger inn $\Delta x = \Delta y = h$:

$$\frac{u_{i+1,j} - 2u_{i,j} + u_{i-1,j}}{h^2} + \frac{u_{i,j+1} - 2u_{i,j} + u_{i,j-1}}{h^2} = 0$$

Ganger ut $h^2$, hvor vi ender opp med 5-punkts Laplace-ligning:

$$u_{i+1,j} - 2u_{i,j} + u_{i-1,j} + u_{i,j+1} - 2u_{i,j} + u_{i,j-1} = 0$$

$$u_{i+1,j} + u_{i-1,j} + u_{i,j+1} + u_{i,j-1} - 4u_{i,j} = 0$$

$$u_{i,j} = \frac{1}{4}(u_{i+1,j} + u_{i-1,j} + u_{i,j+1} + u_{i,j-1})$$

Randsbetingelser diskret:

$$u_{0,j} = \sin(2\pi y_j), \qquad u_{N,j} = \sin(2\pi y_j)$$

$$u_{i,0} = 0, \qquad u_{i,M} = \sin(\pi x_i)$$

Dette gir et lineært likningssystem på formen $Au = b$. Vi løser dette iterativt med Gauss-Seidel, hvor hver indre verdi oppdateres fra naboverdiene inntil løsningen konvergerer:

$$u_{i,j}^{(k+1)} = \frac{1}{4}\left(u_{i+1,j}^{(k)} + u_{i-1,j}^{(k+1)} + u_{i,j+1}^{(k)} + u_{i,j-1}^{(k+1)}\right)$$

Løsningen bestemmes fullstendig av randsbetingelsene.
